```python
import numpy as np
import pandas as pd
from univariate import univar
```

```python
dataset = pd.read_csv("Placement.csv")
dataset
```

```python
quan, qual = univar.quanqual(dataset)
```

```python
quan
```

```python
dataset["salary"].fillna(0, inplace=True)
for column in quan: 
    dataset[column].fillna(dataset[column].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
qual
```

```python
#create and populate table that contains univariate analysis data
descriptive=univar.univariate_table(dataset, quan)
```

```python
descriptive
```

```python
# find the OUTLIERS COLUMNS by comparing min & max with lesser & greater outlier range respectively
lesser_outliers, greater_outliers = univar.find_outliers_columns(quan, descriptive)
```

```python
lesser_outliers
```

```python
greater_outliers
```

```python
#replace the ouliers from the above found columns
#this fundtion doesn't retirn anything. it directly replaces the outliers in the dataset.

univar.replace_outliers(dataset, descriptive, lesser_outliers, greater_outliers)
```
